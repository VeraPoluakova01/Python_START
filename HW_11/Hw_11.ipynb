```python
from sympy.plotting import plot 

from sympy import *
```

Даны две функции: f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241

```python
x = Symbol("x")
```

```python
f  = x**3 - 50*x 
f
```

```python
g = -x**4 + 88*x**2 - 241
g
```

Находим функцию пересечения r

```python
r = f - g
r
```

Находим координаты точек пересечения 

```python
k = solve(r)
```

```python
koord[[x_elem.n().round(2),f.subs(x,x_elem).n().round(2)] for x_elem in k]
```

Стоим графики функций в одной системе координат

```python
plot(f,g)
```

Построить график функции пересечения

```python
plot(r)
```
